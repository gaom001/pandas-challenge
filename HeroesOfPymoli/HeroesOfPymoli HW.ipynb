## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Retrieve data from original dataframe
data_selected1=purchase_data.loc[:,["Purchase ID","Age","Price"]]
```

```python
# Create the bins in which Data will be held, and Create the names for the bins
bins = [4,9,14,19,24,29,34,39,46]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]
```

```python
# Create a column "age ranges" based on the bins
data_selected1["Age Ranges"] = pd.cut(data_selected1["Age"], bins, labels=age_group)
data_selected1
```

```python
# Count the numbers of purchase based on age ranges
df_group_count =data_selected1.groupby(["Age Ranges"]).count()
df_group_count=df_group_count.drop(columns=["Age","Price"])
df_group_count
```

```python
# Calculate the total purchase value in each age range
df_group_total_value =data_selected1.groupby(["Age Ranges"]).sum()
df_group_total_value=df_group_total_value.drop(columns=["Age","Purchase ID"])
df_group_total_value
```

```python
# Calculate average purchase price in each age range and add to the previous table
df_group_total_value["Average Purchase Price"]=df_group_total_value.loc[:,"Price"]/df_group_count.loc[:,"Purchase ID"]
df_group_total_value
```

```python
# Calaulate Avg Total Purchase per Person and add this column to above table
df_group_total_value["Avg Total Purchase per Person"]=df_group_total_value.loc[:,"Price"]/df_age_demographics.loc[:,"Total Count"]
df_group_total_value
```

```python
# Create the final table by merging 
df_merge=pd.merge(df_group_count,df_group_total_value,on="Age Ranges")
df_merge
```

```python
# Reorder and Rename the columns
df_merge=df_merge[["Purchase ID","Average Purchase Price","Price","Avg Total Purchase per Person"]]
df_merge=df_merge.rename(columns={"Purchase ID":"Purchase Count","Price":"Total Purchase Value"})
df_merge
```

```python
# Format "average purchase price","total purchase value" and "avg total purchase per person"
df_merge["Average Purchase Price"]=df_merge["Average Purchase Price"].map("${:.2f}".format)
df_merge["Total Purchase Value"]=df_merge["Total Purchase Value"].map("${:.2f}".format)
df_merge["Avg Total Purchase per Person"]=df_merge["Avg Total Purchase per Person"].map("${:.2f}".format)
df_merge.head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Extact data from orignal dataframe
retrieve_df=purchase_data[["Item ID","Item Name","Price"]]
retrieve_df
```

```python
# Group by Item ID and Item Name, count the numbers, then sort this data by descending order
item_count=retrieve_df.groupby(["Item ID","Item Name"]).count().sort_values(["Price"],ascending=False)
item_count
```

```python
# Group by Item ID and Item Name, list the price for each item
item_price=retrieve_df.groupby(["Item ID","Item Name"]).max()
item_price
```

```python
# Left Joining above two tbales (on index) 
item_df=pd.merge(item_count,item_price,left_index=True,right_index=True,how="left")
item_df
```

```python
# Calculate the total purchase value for each item and add this column into the table
item_df["Total Purchase Value"]=item_df.loc[:,"Price_x"]*item_df.loc[:,"Price_y"]
item_df
```

```python
# Sort the purchase count in descending order, show only top 5 items
item_df.sort_values(["Price_x"],ascending=False)
item_df.head(5)
```

```python
# Give the displayed data cleaner headers
most_popular_item_df=item_df.rename(columns={"Price_x":"Purchase Count","Price_y":"Item Price"})
most_popular_item_df=most_popular_item_df.head(5)
most_popular_item_df
```

```python
# Table formating
most_popular_item_df["Item Price"]=most_popular_item_df["Item Price"].map("${:.2f}".format)
most_popular_item_df["Total Purchase Value"]=most_popular_item_df["Total Purchase Value"].map("${:.2f}".format)
most_popular_item_df.head()
```
